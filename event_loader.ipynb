```python
# Load NeuroKit and other useful packages
import neurokit2 as nk
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Download data
data = nk.data("bio_eventrelated_100hz")
```

```python
print(type(data))
print(data.shape)
print(data)
print(data['Photosensor'].value_counts())
plt.plot(data["Photosensor"])
```

```python
print(data['Photosensor'][ 1000: 1024+500].values)
plt.plot(data["Photosensor"][ 9220: 9224+500])
```

```python
condition_list = ["Negative", "Neutral", "Neutral", "Negative"]
```

```python
# Find events
events = nk.events_find(data["Photosensor"], threshold_keep='below', event_conditions=condition_list)
events
```

```python
# Plot the location of event with the signals
plot = nk.events_plot(events, data)
```
